### 1. Sequence Validation Function
* making sure sequence doesn't contain any unexpected characters
* `validate seq` with sequence for short function accepts a string
* it then creates a temporary string which is upper case
* we then just loop through each character in that string to make sure it is one of these four characters
* if at least one character does not match break out of the loop and we can return false because we're not interested in working with a faulty string

```python
# checking the sequence to make sure it's a DNA String
def validateSeq(dna_seq):
  tmpseq = dna_seq.upper()
  for nuc in tmpseq:
    if nuc not in Nucleotides:
      return False
  return tmpseq
```

```python
# testing function (add/remove X, change cases)
rndDNAstr = "ATTGCAatCCGT"
print(validateSeq(rndDNAstr))
```

```python
# generating random DNA string
import random

randDNAstr = ''.join([random.choice(Nucleotides)
for nuc in range(20)])

print(validateSeq(randDNAstr))
```

### 2. Nucleotide Frequency Function

* write a function that returns the proportional representation of the four bases (ATGC)

* Nucleotide frequency is a fundamental metric in bioinformatics,
* Applications:
  1. identify gene structures
  2. detect evolutionary patterns
  3. classify species based on their genetic signatures.

```python
def countNucFreq(dna_seq):
  tmpFreqDict = {"A": 0, "T":0, "C":0, "G":0}
  for nuc in dna_seq:
    tmpFreqDict[nuc] += 1
  return tmpFreqDict, dna_seq

print(countNucFreq(randDNAstr))
```

```python
# applying collections module in py for one string code
import collections

def countNuqFreq(dna_seq):
  return dict(collections.Counter(dna_seq))

print(countNuqFreq(randDNAstr))
```

#### Strand Complement Function
* present result in the 5' - 3' orientation
* requires [::-1] string reversal
* provide the reverse complement

```python
complements = {"A": "T", "T": "A", "C": "G", "G": "C"} # pairing rules

# Looping through every character and building a new list
def complementStr(dna_seq):
  complement_strand = []
  for nuc in dna_seq:
    complement_strand.append(complements[nuc])
    complement = "".join(complement_strand) # Join the list back into a single string
    revComp = complement[::-1] # Reverse the string
  return revComp
```

Given: A string s
 of length at most 200 letters and four integers a
, b
, c
 and d
.

Return: The slice of this string from indices a
 through b
 and c
 through d
 (with space in between), inclusively. In other words, we should include elements s[b]
 and s[d]
 in our slice.

```python
def slicerProMax(str:str, a:int, b:int, c:int, d:int):
  b += 1
  d += 1
  string = str[a:b], str[c:d]
  return string
```

```python
# odds

def SumOdds(n, n2):
  return sum(num for num in range (n, n2 +1) if num % 2 != 0)

print(SumOdds(4820, 9588))
```

```python
# Enen numbered File lines
# Open the source file to read and a new file to write the output
# Return only the even numbered lines in the file


def getEven (file):
  with open(file, "r") as infile, open("output.txt", "w") as outfile:
    for line_num, line in enumerate(infile, 1):
        if line_num % 2 == 0:
            outfile.write(line)

  return outfile
```
