### 1. Sequence Validation Function
* making sure sequence doesn't contain any unexpected characters
* `validate seq` with sequence for short function accepts a string
* it then creates a temporary string which is upper case
* we then just loop through each character in that string to make sure it is one of these four characters
* if at least one character does not match break out of the loop and we can return false because we're not interested in working with a faulty string

```python
# checking the sequence to make sure it's a DNA String
def validateSeq(dna_seq):
  tmpseq = dna_seq.upper()
  for nuc in tmpseq:
    if nuc not in Nucleotides:
      return False
  return tmpseq
```

### 2. Nucleotide Frequency Function

* write a function that returns the proportional representation of the four bases (ATGC)

* Nucleotide frequency is a fundamental metric in bioinformatics,
* Applications:
  1. identify gene structures
  2. detect evolutionary patterns
  3. classify species based on their genetic signatures.

```python
def countNucFreq(dna_seq):
  tmpFreqDict = {"A": 0, "T":0, "C":0, "G":0}
  for nuc in dna_seq:
    tmpFreqDict[nuc] += 1
  return tmpFreqDict, dna_seq

print(countNucFreq(randDNAstr))
```

#### Strand Complement Function
* present result in the 5' - 3' orientation
* requires [::-1] string reversal
* provide the reverse complement

```python
complements = {"A": "T", "T": "A", "C": "G", "G": "C"} # pairing rules

# Looping through every character and building a new list
def complementStr(dna_seq):
  complement_strand = []
  for nuc in dna_seq:
    complement_strand.append(complements[nuc])
    complement = "".join(complement_strand) # Join the list back into a single string
    revComp = complement[::-1] # Reverse the string
  return revComp
```

```python
def slicerProMax(str:str, a:int, b:int, c:int, d:int):
  b += 1
  d += 1
  string = str[a:b], str[c:d]
  return string
```

```python
def fibonacci(n):     # add 1 to n if you're interested in the sum of last two numbers(like 6th fibonacci num is 8)
  if n == 0 : return [0]
  if n == 1 : return [0, 1]
  seq = [0, 1]
  while len(seq) < n:
    next_num= seq[-1] + seq[-2]
    seq.append(next_num)
  return seq[-1]

print(fibonacci(23))
```

```python
def rabbits(n:int, k:int) -> int:
  if n == 0: return 0
  seq = [1, 1]
  while len(seq) <= n:
    next_num = seq[-1] + (k*seq[-2])
    seq.append(next_num)
  return seq[n-1]

print(rabbits(31, 4))
```

Recall the definition of the Fibonacci numbers from “Rabbits and Recurrence Relations”, which followed the recurrence relation Fn=Fn−1+Fn−2
 and assumed that each pair of rabbits reaches maturity in one month and produces a single pair of offspring (one male, one female) each subsequent month.

Our aim is to somehow modify this recurrence relation to achieve a dynamic programming solution in the case that all rabbits die out after a fixed number of months. See Figure 4 for a depiction of a rabbit tree in which rabbits live for three months (meaning that they reproduce only twice before dying).

Given: Positive integers n≤100
 and m≤20
.

Return: The total number of pairs of rabbits that will remain after the n
-th month if all rabbits live for m
 months

```python
# wrong
def mortal_rabbits(n: int, m: int) -> int:
    # Track population by age; start with 1 newborn pair
    ages = [1] + [0] * (m - 1)

    for _ in range(1, n):
        # Newborns = sum of all mature rabbit pairs (index 1 to m-1)
        # Shift generations right (aging them); oldest drop off (dying)
        ages = [sum(ages[1:])] + ages[:-1]

    return sum(ages)

# Example Usage
print(mortal_rabbits(6, 3))   # Output: 4
print(mortal_rabbits(84, 16)) # Output: 24715535359411985
```

```python
def solve_mortal_rabbits(n: int, m: int) -> int:
    # Track the exact count of rabbit pairs at each specific age
    # index 0 is newborns, index 1 is 1-month-olds, ..., index m-1 is the oldest rabbits
    age_groups = [0] * m

    # Month 1: We start with exactly 1 newborn pair
    age_groups[0] = 1

    # Simulate the passage of time month by month from Month 2 up to Month n
    for _ in range(1, n):
        # 1. Any rabbit pair that is 1 month or older (index 1 to m-1) reproduces
        newborns_produced = sum(age_groups[1:])

        # 2. Age all rabbits by shifting them up 1 month.
        # The rabbits at the very last index (age_groups[m-1]) are naturally
        # dropped because they have reached their maximum lifespan m.
        next_generation = [0] * m
        for i in range(1, m):
            next_generation[i] = age_groups[i - 1]

        # 3. Place the brand new babies into the newborn slot (index 0)
        next_generation[0] = newborns_produced

        # 4. Update our active tracker for the next iteration
        age_groups = next_generation

    # The total population remaining is the sum of all surviving age groups
    return sum(age_groups)

# --- VERIFICATION CHECKS ---
if __name__ == "__main__":
    # 1. Rosalind Sample Test (n=6, m=3) -> Expected Output: 4
    print(f"Sample (6, 3): {solve_mortal_rabbits(6, 3)}")

    # 2. Your first test case (n=84, m=16) -> Expected Output: 158240333533794817
    print(f"Test (84, 16): {solve_mortal_rabbits(84, 16)}")

    # 3. Your second test case (n=86, m=16) -> Expected Output: 414110157126129233
    print(f"Test (88, 19): {solve_mortal_rabbits(86, 16)}")
```
